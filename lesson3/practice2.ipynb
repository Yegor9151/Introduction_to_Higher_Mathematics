# 1. Задание (в программе)

Нарисуйте график функции:   для некоторых (2-3 различных) значений параметров k, a, b

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x, k=1, a=0, b=0):
    return k * np.cos(x - a) + b

x = np.linspace(-10, 10, 1000)
y = func(x)

y1 = func(x, k=2)
y2 = func(x, a=2)
y3 = func(x, b=2)

plt.figure(figsize=(8, 4))

plt.plot(x, y, label='default', linestyle='--')
plt.plot(x, y1, alpha=0.3, label="k=2")
plt.plot(x, y2, alpha=0.3, label="a=2")
plt.plot(x, y3, alpha=0.3, label="b=2")

plt.legend()
plt.grid()
```
