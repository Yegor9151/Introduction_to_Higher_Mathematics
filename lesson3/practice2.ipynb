# 1. Задание (в программе)

Нарисуйте график функции:   для некоторых (2-3 различных) значений параметров k, a, b

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x, k=1, a=0, b=0):
    return k * np.cos(x - a) + b

x = np.linspace(-10, 10, 1000)
y = func(x)

y1 = func(x, k=2)
y2 = func(x, a=2)
y3 = func(x, b=2)

plt.figure(figsize=(8, 4))

plt.plot(x, y, label='default', linestyle='--')
plt.plot(x, y1, alpha=0.3, label="k=2")
plt.plot(x, y2, alpha=0.3, label="a=2")
plt.plot(x, y3, alpha=0.3, label="b=2")

plt.legend()
plt.grid()
```

1. Напишите код, который будет переводить полярные координаты в декартовы.

```python
def get_decards(r, alpha):
    x = r*np.cos(alpha)
    y = r*np.sin(alpha)
    return x, y

x, y = get_decards(1, 30)
x, y
```

2. Напишите код, который будет рисовать график окружности в полярных координатах.

```python
def oval(a, b, x0=0, y0=0):
    x = np.linspace(-a, a, 1000)
    y = (b**2 - ((x * b)**2) / a**2)**0.5
    
    return x+x0, y+y0, -y+y0

x, y1, y2 = oval(4, 4)
plt.polar(x, y1)
plt.polar(x, y2)
```

3. Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.

```python
def line(x1=-5, x2=5, k=1):
    x = np.linspace(x1, x2, 1000)
    y =  k * x
    
    return x, y

x, y = line()
plt.polar(x, y)
```
